***2- Data Analysis:***

**1- Perform basic statistical analysis using Numpy to summarize the data**

a- Calculate the mean, median, and standard deviation of relevant numeric columns

```python
numeric_columns = [
    'extended', 'latitude', 'longitude', 'specificity', 'vicinity',
    'doubtterr', 'multiple', 'nperps', 'nperpcap', 'claime_dresp',
    'targsubtype1', 'natlty1', 'guncertain1', 'num_killed', 'nkillus',
    'num_terrorist_killed', 'num_wounded', 'nwoundus', 'num_terrorist_wounded',
    'propextent', 'ishostkid', 'ransom_demanded'
]

stats = {}
for col in numeric_columns:
    stats[col] = {
        'mean': np.mean(cleaned_df[col].dropna()),
        'median': np.median(cleaned_df[col].dropna()),
        'std_dev': np.std(cleaned_df[col].dropna())
    }

stats_df = pd.DataFrame(stats)
stats_df.head()
```

b- Identify the most frequent values in categorical columns.

```python
def get_most_frequent(values):
    values = np.array(values)
    unique, counts = np.unique(values, return_counts=True)
    max_count_index = np.argmax(counts)
    return unique[max_count_index]

categorical_columns = [
    'country', 'region', 'provstate', 'city', 'attack_type', 'target_type',
    'target_subtype', 'corp1', 'target1', 'nationality_of_victom',
    'organisation', 'motive', 'weaptype1', 'weapon_type', 'weapsubtype1_txt',
    'weapdetail', 'propextent_desc', 'propcomment', 'scite1', 'scite2','dbsource'
]


frequent_values = {}
for col in categorical_columns:
    frequent_values[col] = get_most_frequent(cleaned_df[col].dropna()) 

frequent_values_df = pd.DataFrame(frequent_values, index=['Most Frequent'])
frequent_values_df.head()
```

b- Identify trends over time (e.g., number of attacks per year)

```python
attacks_per_year = cleaned_df.groupby('year').size()
attacks_per_year_df = attacks_per_year.reset_index(name='num_attacks')
attacks_per_year_df
```

```python
#here i want to know the max and the min attacks happened in which year
max_attacks_year = attacks_per_year.idxmax()
max_attacks_count = attacks_per_year.max()
print(f"Max Attacks happen in year: {max_attacks_year} with ({max_attacks_count} attacks)")
min_attacks_year = attacks_per_year.idxmin()
min_attacks_count = attacks_per_year.min()
print(f"Max Attacks happen in year: {min_attacks_year} with ({min_attacks_count} attacks)")
```

```python
#here i want to know the changes in year 
yearly_changes = attacks_per_year.diff().dropna()

print(yearly_changes)
```
